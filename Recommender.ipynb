# Setup

### Library import
Import all the required Python libraries.

It is a good practice to organize the imported libraries by functionality, as shown below.

```python
#local libraries for system runtime
import os
import math
import warnings
warnings.filterwarnings('ignore')

#Data manipulation
import numpy as np
import pandas as pd
import glob
import pickle

#Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


#Image preprocessing
import cv2
from keras.preprocessing.image import ImageDataGenerator

#Modelling
from keras import models, Model

#model evaluation
from sklearn.metrics import accuracy_score,confusion_matrix
```

### Parameter definitions

```python
# Set all relevant parameters for the notebook. By convention, parameters are uppercase, while all the other variables follow Python's guidelines.
BATCH_SIZE = 32
TARGET_SIZE = (96, 96)
SEED = 1234
RANDOM = 42
```

### Data import

```python
df = pd.read_csv("dataset/rec_style.csv") #import the dataframe dataset containing the image features and class
image_dir = "dataset/images/"
sample= "dataset/test/"
ris_model = models.load_model('saved_model/ris_val_0.3.h5') #import model
#import image targets
pickle_in = open('saved_model/labels.pickle', 'rb')
labels = pickle.load(pickle_in)  
```

```python
#Creating a DataFrame by passing a dict of objects
features = pd.DataFrame( columns=['filename'])
features
```

#### Data Preparation

```python
#fill id column with image filename
img_index = ""
image_data = glob.glob(sample+"*.jpeg")
for i, filename in enumerate(sorted(image_data)):
    image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    img_index = image
    plt.imshow(image)
    filename=filename.replace("dataset/test/","") #strip off filename path
    features = features.append({"filename":filename},ignore_index=True) #append filename to id column
features.head()
```

### Data Prepreprocessing

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_dataframe(features, directory=sample, x_col='filename', y_col=None,target_size=TARGET_SIZE, color_mode='rgb',
    class_mode=None, batch_size=BATCH_SIZE, shuffle=False,
    seed=SEED)
test_set.reset()
```

### Model Predicted

```python
def ShowImage(test_set):
    predictions = ris_model.predict_generator(test_set)
    pred_class = np.argmax(predictions,axis=1)
    for y in pred_class:
        result = list(labels.keys())[list(labels.values()).index(y)]
        return predictions, result
```

```python
sample_image,predicted_class=ShowImage(test_set)
```

### Recommender Systems

```python
similar=[]
for i in range(0,len(df)): 
    if(df["color"][i]==predicted_class):
        similar.append(int(df['filename'][i].replace('.png','')))
            
print(len(similar))
```

### Append all similar products and their IDs into arrays

```python
similar_images=[]
similar_IDs=[]

for imageID in similar:
    Image_path=image_dir+str(imageID)+".png"
    image = cv2.imread(Image_path,cv2.COLOR_BGR2RGB)
    similar_images.append(image)
    similar_IDs.append(imageID)
        
similar_images = np.array(similar_images).reshape(-1,96,96,3)
similar_images = similar_images/255
```

### Calculating Euclidean Distance between two grayscale images

```python
def calculateDistance(i1, i2):
    return math.sqrt(np.sum((i1-i2)**2))
```

### Display the sample image

```python
plt.imshow(img_index)
plt.title("sample_id")
```

### Display one of the similar images

```python
plt.imshow(similar_images[2])
plt.title(similar_IDs[2])
```

### Display the Euclidean distance between the sample image and the above similar image

```python
print("Vector Distance bwteen the sample and the above similar image is: ",calculateDistance(sample_image,similar_images[2]))
```

## Calculate the Euclidean distance between the sample image and all the similar images

#### Get the indices of the best 10 similar images

```python
distance_list=[]
for i in range (0, len(similar_images)):
    distance_list.append(calculateDistance(sample_image,similar_images[i]))

sorted_distance_list=distance_list.copy()
#print(distance_list)
sorted_distance_list.sort()

best10=sorted_distance_list[0:11]
print(best10)
index_distance=[]
for i in range (0, len(best10)-1):
    if(best10[i]!=best10[i+1]):
        index_distance.append(distance_list.index(best10[i]))
print(index_distance)
```

#### Display the best 10 similar images as recommended products

```python
final_images=[]

for i in range(0,len(index_distance)):
    print(similar_IDs[i])
    Image_path=image_dir+str(similar_IDs[index_distance[i]])+".png"
    src = cv2.imread(Image_path)
    image = cv2.cvtColor(src, cv2.COLOR_BGR2RGB) 
    final_images.append(image)
    
```

```python
plt.imshow(img_index)
plt.title("sample_id")
fig = plt.figure(figsize=(25, 25))
  
rows = 5
columns = 2

for i in range(0,len(final_images)):
    
    fig.add_subplot(rows, columns, i+1)

    plt.imshow(final_images[i])
    plt.title(similar_IDs[index_distance[i]])
    plt.axis('off')
```
